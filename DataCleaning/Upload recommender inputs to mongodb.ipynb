### Dictionary users: recipes -> ratings (just in case)

```python
import csv

def  from_csv_to_dict(filename):

    class AutoVivification(dict):
        """Implementation of perl's autovivification feature."""
        def __getitem__(self, item):
            try:
                return dict.__getitem__(self, item)
            except KeyError:
                value = self[item] = type(self)()
                return value

    d = AutoVivification()
    with open(filename, 'r') as f:
        reader = csv.reader(f, delimiter=',')
        next(reader)        # skip the header
        for row in reader:
            d[row[0]][row[1]] = row[2] 
            
    return d
```

```python
ratings_dic_json = from_csv_to_dict("ratings.csv")

import json
with open("users-recipes-ratings.json", 'w') as outfile:
    json.dump(ratings_dic_json, outfile)
```
