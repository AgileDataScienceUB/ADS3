```python
users=len(matrixR.columns)
users
```

```python
def distance_recipes(recipe1,recipe2,matrix):
    return (np.sum(matrix.loc[recipe1].values*matrix.loc[recipe2].values))/np.sum(matrix.loc[recipe1].values)

def average_rating(recipe1,matrixR):
    k=0
    #r=0
    for i in range (0,users):
        if matrixR.loc[recipe1][i]!=0:
            k=k+1  
            #r=r+matrixR.loc[recipe1][i]
    #return r/k
    return np.sum(matrixR.loc[recipe1].values)/k
```

```python
def rec_user(recipe1):
    aux = pd.DataFrame(index = matrix.index)
    aux['dist'] = 0.0
    #aux['av_rating'] = 0.0
    for rec in matrix.index:
        aux.loc[rec]['dist'] = distance_recipes(recipe1,rec,matrix)
    df_return = aux[(aux["dist"]!=1)].sort_values(by=["dist"],ascending=False)[0:10]
    #for rec in df_return.index:
    #    aux.loc[rec]['av_rating'] = np.round(average_rating(rec,matrixR), decimals=2)
    #df_return['av_rating'] = aux['av_rating']
    return np.array(df_return.index)
```

```python
rec_user("spinach-cheese-rolls")
```

```python
df_result = pd.DataFrame(columns=['recipe', 'average_rating', 'most_similar1', 'most_similar2', 'most_similar3', 'most_similar4', 'most_similar5', 'most_similar6', 'most_similar7', 'most_similar8', 'most_similar9', 'most_similar10'])
aux = pd.DataFrame(index = matrix.index)

for idx,rec in enumerate(matrix.index):
    #print(rec)
    avv = np.round(average_rating(rec,matrixR), decimals=2)
    fff = rec_user(rec)
    df_result.loc[idx] = [rec, avv, fff[0], fff[1], fff[2], fff[3], fff[4], fff[5], fff[6], fff[7], fff[8], fff[9]]

df_result = df_result.set_index('recipe')
df_result
```
