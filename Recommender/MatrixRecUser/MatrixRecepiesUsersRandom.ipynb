```python
import numpy as np
import pandas as pd
```

```python
ingredients = pd.read_csv('Source1Ingredients.csv')
```

```python
ids = np.unique(ingredients['id'])
```

```python
user = ['user_'+str(i) for i in range(1, 10001, 1)]
```

```python
random = np.full((len(ids), len(user)), np.nan)
```

```python
from tqdm import tqdm
```

```python
for i in tqdm(range(len(ids))):
    for j in range(len(user)):
        if np.random.randint(0,11) > 8:
            random[i,j] = np.random.randint(4,11)
```
