```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
ing_1 = pd.read_csv('Source1Ingredients.csv')
ing_2 = pd.read_csv('Source2Ingredients.csv')
```

```python
ids_1 = np.unique(ing_1['id'])
ids_2 = np.unique(ing_2['id'])
```

```python
ids_1.shape, ids_2.shape
```

```python
ids = np.concatenate((ids_1, ids_2), axis = 0)
ids.shape
```

```python
ratings = {}
for _id in tqdm(ids):
    ratings[_id] = np.random.randint(3,5)
```

```python
ratings
```

```python
user = ['user_'+str(i) for i in range(0, 66, 1)]
```

```python
#mu, sigma = 0, 0.1 # mean and standard deviation
#s = np.random.normal(mu, sigma, 1000)
```

```python
random = np.full((len(ids), len(user)), np.nan)
for i in tqdm(range(len(ids))):
    for j in range(len(user)):
        if np.random.randint(0,11) > 7:
            random[i,j] = min(max(int(np.random.normal(ratings[ids[i]], min(1, 5 - ratings[ids[i]]))), 1), 5)
```

```python
matrix = pd.DataFrame(random, index=ids, columns=user)
```

```python
matrix
```

```python
print(np.mean(matrix[:3], axis = 1))
print(np.unique(matrix[:3], return_counts=True))
```
