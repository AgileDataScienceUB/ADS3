```python
input_integredient=['spaghetti','olive oil']
```

```python
found_int=db.collection.aggregate(
    [ 
        { "$match": { "list_integredients": { "$exists": True } } }, 
        { "$redact": { 
            "$cond": [ 
                { "$gte": [ 
                    { "$size": { "$setIntersection": [ "$list_integredients", input_integredient ] } }, 
                    1
                ]},
                "$$KEEP", 
                "$$PRUNE" 
            ]
        }}
    ]
)
```

```python
print(list(found_int))
```
