```python
ingredients = ingredients[~ingredients['ingredient'].str.contains("<em>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<h5>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<h4>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<img", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<strong>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<a href", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("&nbsp", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<i>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<small>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<b>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<br>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<br", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("</br>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("</p>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("</ul>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<ul", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<li>", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<Strong", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<bold", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("</bold", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("</str", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<a ", na=False)]
ingredients = ingredients[~ingredients['ingredient'].str.contains("<span", na=False)]
```

```python
nRecepies = 13244
nMaxIngridients = 42

columnsData = []
for i in range(1, 43):
    columnsData.append('ingredient_'+str(i))
    
data = np.empty((nRecepies, nMaxIngridients,))
data[:] = np.NAN
data.shape

data = pd.DataFrame(data, columns=columnsData )
```

```python
ids = np.unique(ingredients['id'])
ids = pd.DataFrame(ids, columns=['id'])
```

```python
_, c = np.unique(ingredients['id'], return_counts = True)
counts = pd.DataFrame(c, columns=['ningredients'])
```

```python
dataIngr = pd.concat([ids, data, counts],axis =1 )
```

```python
dataIngr.shape, data.shape, ids.shape, counts.shape
```

```python
ingredients.shape
```

```python
from tqdm import tqdm
old_recepie = None
countIngridient = 1
for recepie, ingridient in tqdm(ingredients.values):
    if old_recepie != recepie:
        old_recepie = recepie
        countIngridient = 1
    else:
        countIngridient += 1
    
    column = "ingredient_"+str(countIngridient)
    dataIngr.loc[dataIngr['id'] == recepie, column] = ingridient
```

```python
dataIngr
```

```python
def clean_ingredients(integredient):
    l_ingr_name=integredient.split(",")
    clean_ingr_name=l_ingr_name[0]
    if(len(l_ingr_name)>1):
        for i in range(1,len(l_ingr_name)):
            clean_ingr_name+="("+str(l_ingr_name[i]).strip()+")"
    return clean_ingr_name
```

```python
dataIngr
```

```python
columns=list(dataIngr)
columns.remove("id")
columns.remove('ningredients')

dataIngr=dataIngr.fillna("")
dataIngr[columns]=dataIngr[columns].applymap(lambda x: clean_ingredients(x))
dataIngr
```

```python
#Join all ingredient to a list of ingredients
value=list(dataIngr.drop(["id",'ningredients'],axis=1).fillna('').values.tolist())
dataIngr['list_ingredients']=value
dataIngr['list_ingredients']=dataIngr['list_ingredients'].apply(lambda x:list(filter(None, x)) )
dataIngr.list_ingredients.loc[0]
```

```python
#Now we keep the cleaned Ingredient dataset 
dataIngr=dataIngr[["id","list_ingredients","ningredients"]]
dataIngr
```

```python
result = recepies.merge(dataIngr, how="left",on="id")
```
